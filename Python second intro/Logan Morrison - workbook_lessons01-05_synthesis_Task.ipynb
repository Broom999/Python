## Section F — Putting it all together

### Problem 10 — Profile flag generator (capstone)

Write a program that builds a small "profile summary" using everything from Lessons 01–05. Ask the user for:

- `firstName` (str)
- `age` (int)
- `gpa` (float)
- `favoriteSubject` (str)

Then build these booleans:

- `isTeen` — `13 <= age <= 19` (chained comparison)
- `isAdult` — age is 18 or over
- `isHonorRoll` — gpa is 3.5 or higher
- `lovesMath` — favoriteSubject is `"math"` (case-insensitive — use `.lower()`)

Print a clean summary using f-strings. Show `gpa` to **2 decimal places**. Example:

```
------------------------------
        PROFILE
------------------------------
Name: Maria
Age: 17
GPA: 3.80
Favorite subject: Math

Flags:
  isTeen:       True
  isAdult:      False
  isHonorRoll:  True
  lovesMath:    True
------------------------------
```

Add a header comment block at the top of your code with `Program:`, `Author:`, `Date:`, and `Purpose:` (just like Lesson 01).

````python
# Your code here
name = input("name: ")
age = int(input("age: "))
gpa = float(input("gpa: "))
fav = input("Favorite subject: ")
teen = bool(13 <= age <= 19)
adult = bool(age >= 18)
honor = bool(gpa >= 3.8)
matt = "math"
math = bool(fav.lower() == matt)
print(f"------------------------------")
print(f"        PROFILE")
print(f"------------------------------")
print(f"Name: {name}")
print(f"Age: {age}")
print(f"GPA: {gpa}")
print(f"Favorite subject: {fav}")

print(f"Flags:")
print(f"  isTeen:       {teen}")
print(f"  isAdult:      {adult}")
print(f"  isHonorRoll:  {honor}")
print(f"  lovesMath:    {math}")
print(f"------------------------------")
````
