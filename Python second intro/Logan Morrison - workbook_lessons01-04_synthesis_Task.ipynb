### Problem 9 — Discount calculator

Ask the user for an `originalPrice` (float) and a `discountPercent` (float, like `25` for 25%). Calculate:

- the dollar amount saved
- the final sale price

Print both using an f-string with **2 decimal places**. Example:

```
Original price: 80
Discount %: 25
You save $20.00
Sale price: $60.00
```

````python
# Your code here
oP = float(input("Original price: "))
dP = float(input("Discount: "))
dpp = dP / 100
sale = oP - (oP * dpp)
save = oP - sale
dpp = dP * 100
print(f"Origin: ${oP} Discount {dP}% you save ${save} new price {sale}")
````

### Problem 10 — BMI calculator

Ask the user for their weight in **pounds** (float) and height in **inches** (float). Calculate BMI using:

$$BMI = \frac{weight}{height^2} \times 703$$

Print the result rounded to **1 decimal place**. Example:

```
Weight (lbs): 150
Height (in): 68
Your BMI is 22.8
```

Hint: use `**` for `height^2`.

````python
# Your code here
wei = int(input("How many pounds to do you weigh: "))
hei = int(input("what is your height in inches: "))
bmi = (wei/(hei ** 2)) * 703
print(f"{wei} pounds {hei} inches {bmi:.1f} bmi")
````

## Section F — Putting it all together

### Problem 14 — Coin counter

Ask the user how many of each coin they have (all `int`):

- `quarters`
- `dimes`
- `nickels`
- `pennies`

Calculate the total value in **dollars** (a float). Print a summary using f-strings, with the dollar total formatted to **2 decimal places**.

Example:

```
Quarters: 3
Dimes: 4
Nickels: 1
Pennies: 7
You have 3 quarters, 4 dimes, 1 nickels, and 7 pennies.
Total: $1.27
```

Hint: 25 cents + 25 cents + 25 cents + ... is easier as `quarters * 25`. Add up all the cents, then divide by 100 to get dollars.

````python
# Your code here
quarter = int(input("How many quarters do you have: "))
dime = int(input("How many dimes do you have: "))
nickel = int(input("How many nickels do you have: "))
penny = int(input("How many pennies do you have: "))
Qu = (quarter / 100) * 25
Di = (dime / 100) * 10
Ni = (nickel / 100) * 5
Pe = (penny / 100)
print(f"you have {quarter} quarters, {dime} dimes, {nickel} nickels, and {penny} pennies")
money = Qu + Di + Ni + Pe
print(f"Total: ${money}")
````

### Problem 15 — Mini receipt generator (capstone)

Write a small program that asks the user for:

- `itemName` (str)
- `quantity` (int)
- `unitPrice` (float)
- `taxPercent` (float) — like `7.25` for 7.25%

Then calculate:

- `subtotal = quantity * unitPrice`
- `tax = subtotal * (taxPercent / 100)`
- `total = subtotal + tax`

Print a clean receipt using f-strings, with all dollar amounts to **2 decimal places**. Example:

```
------------------------------
         RECEIPT
------------------------------
Item: Coffee
Quantity: 3
Unit price: $4.50
Subtotal: $13.50
Tax (7.25%): $0.98
Total: $14.48
------------------------------
```

Add a header comment block at the top of your code with `Program:`, `Author:`, `Date:`, and `Purpose:` (just like Lesson 01).

````python
# Your code here
itemName = input("What is the name of the item you bought: ")
quantity = int(input("How much: "))
unit = float(input("How much did it cost: "))
taxP = float(input("What is the percent tax: "))
subtotal = quantity * unit
tax = subtotal * (taxP / 100)
total = subtotal + tax
print(f"------------------------------")
print(f"           RECEIPT")
print(f"------------------------------")
print(f"Item: {itemName}")
print(f"Quantity: {quantity}")
print(f"Unit price: ${unit}")
print(f"Subtotal: ${subtotal}")
print(f"Tax ({taxP}%): ${tax:.2f}")
print(f"Total: ${total:.2f}")
print(f"------------------------------")
````
