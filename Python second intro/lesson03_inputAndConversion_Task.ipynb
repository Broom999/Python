## Exercise 3 — Age in 10 years

Ask the user their age, **convert it to an int**, add 10, and print the result. Example:

```

```

````python
# Your code here
age = int(input("How old are you? "))
age_plus = age + 10
print(age)
print("In 10 years you will be ", age_plus)
````
