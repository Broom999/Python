```python
# Program: Receipt calculator
# Author: Ryan
# Purpose: compute total, tax, tip, and a grand total for a meal

priceText = input("Meal price: $")
price = float(priceText)

taxRate = input("What is the tax rate: ")
tax = float(taxRate)

tipRate = input ("What percentage of your meal would you like to tip: (Please use decimals instead of percentages) ")
tip = float(tipRate)
total = price + (price * tax) + (tip * price)
print(f"The price of your meal is ${price}, the tax rate is {tax}, you asked to tip {tip * 100:.0f}%, and the total cost of your meal is ${total:.2f}")
```
